```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://pythonhow.com/example.html")
c=r.content
print(c)
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
print(soup.prettify())
```

```python
p=soup.findAll("p")
```

```python
h=soup.find_all("h1")
```

```python
p,h
```

```python
cities=soup.findAll("h1",{"class":"cities"})[0] # This finds all of the h1 tags in the html file witht he class of cities.
# I am keeping this so I can see the syntax for later.
```
